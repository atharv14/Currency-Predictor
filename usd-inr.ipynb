```python
x = INR_data[["Open", "High", "Low"]]

y = INR_data["Close"]
x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1, 1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
```

```python
INR_pred_data = pd.DataFrame(data={"Predicted Rate": ypred.flatten()})
print(INR_pred_data.head())
```
